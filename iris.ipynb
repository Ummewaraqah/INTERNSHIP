Task 1: Exploring and Visualizing a Simple Dataset

OBJECTIVE Understand how to read, summarize, and visualize a dataset.

Dataset: Iris Datase


1. Load the dataset using pandas library

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target # Add the target variable (species) to the DataFrame
df['species'] = iris.target_names[iris.target] # Add human-readable species names

# Display the first few rows of the dataset
print("--- First 10 Rows of the Iris Dataset ---")
print(df.head(10))
```

2. Display dataset structure using .shape, .columns, and .head().

```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame from the data and feature names
# sklearn datasets typically store features in .data and names in .feature_names
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Now, display the structure:

# .shape
print("Shape of the dataset:", iris_df.shape)

# .columns
print("\nColumns of the dataset:", iris_df.columns)

# .head()
print("\nFirst 5 rows of the dataset:\n", iris_df.head())
```

3. Create basic visualizations:

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

# Load dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

# 1. Scatter Plot (Sepal Length vs. Petal Length)
sns.scatterplot(data=df, x="sepal length (cm)", y="petal length (cm)", hue="species")
plt.title("Sepal vs. Petal Length by Species")
plt.show()

# 2. Box Plot (Feature Distributions)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop(columns="species"))
plt.title("Distribution of Iris Features")
plt.xticks(rotation=45)
plt.show()
# 3. Histogram (Distribution of Sepal Length)
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x="sepal length (cm)", hue="species", kde=True, bins=20)
plt.title("Distribution of Sepal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()

# 4. Box Plot (Spread & Outliers for All Features)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop(columns="species"), orient="h")
plt.title("Box Plot of Iris Features (Outliers & Spread)")
plt.xlabel("Value (cm)")
plt.show()

# 5. Grouped Box Plot (e.g., Petal Width by Species)
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="species", y="petal width (cm)")
plt.title("Petal Width Distribution by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.show()
```
